load the data analysis libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

get the data/ generate data

```python
rainfall = [17, 10, 10, 10, 26, 59]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
```

```python
plt.style.available
```

```python
plt.style.use('ggplot') # theme
```

visualize the data

```python
plt.bar(months, rainfall)
```

```python
plt.pie(
    rainfall, 
    labels=months, 
    autopct='%.1f%%',
)
plt.show()
```

```python
# donut
plt.pie(
    rainfall, 
    labels=months, 
    autopct='%.1f%%',
    wedgeprops={'width': .4, 'edgecolor': '#6666ff', 'linewidth': 2},
    pctdistance=.8
)
plt.show()
```

```python
plt.boxplot(rainfall, 
            vert=False, 
            showmeans=True, 
            meanline=True, 
            meanprops={'color': 'red'}
            )
plt.show()
```
